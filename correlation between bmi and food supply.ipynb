```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\bgane\Desktop\TOPICS DATA\data\merged_data_all.csv"  
data = pd.read_csv(file_path)

# Calculate the correlation between BMI and Food Supply Value
correlation = data['FactValueNumeric'].corr(data['Supply Value'])

# Plot the correlation using a scatter plot and regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='FactValueNumeric', y='Supply Value', data=data, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

# Add titles and labels
plt.title(f'Correlation Between BMI and Food Supply Value (r={correlation:.2f})', fontsize=14)
plt.xlabel('BMI (FactValueNumeric)', fontsize=12)
plt.ylabel('Food Supply Value', fontsize=12)
plt.grid(alpha=0.3)

# Show the plot
plt.show()

# Print the correlation coefficient
print(f"Correlation Coefficient: {correlation:.2f}")
```
