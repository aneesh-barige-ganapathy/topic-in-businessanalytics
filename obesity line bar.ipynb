```python

import pandas as pd
import matplotlib.pyplot as plt

 
bmi_data = pd.read_csv(r"C:\Users\bgane\Desktop\TOPICS DATA\obese bmi data.csv")

# Checking for unique regions for grouping
regions = bmi_data['ParentLocation'].unique()

# Filtering data for trend analysis and aggregate by year and region
trend_data = bmi_data.groupby(['Period', 'ParentLocation'])['FactValueNumeric'].mean().reset_index()

# Creating a line graph for obesity prevalence trends by region over time
plt.figure(figsize=(12, 6))
for region in regions:
    region_data = trend_data[trend_data['ParentLocation'] == region]
    plt.plot(region_data['Period'], region_data['FactValueNumeric'], label=region, marker='o')

plt.title('Trends in Obesity Prevalence by Region (2012-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Obesity Prevalence (%)', fontsize=12)
plt.legend(title='Region', fontsize=10)
plt.grid(True)
plt.show()
```
