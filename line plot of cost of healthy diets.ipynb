```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

cost_data_path = r"C:\Users\bgane\Downloads\cost over 5 years.csv"  # Replace with the path to your file
cost_data = pd.read_csv(cost_data_path)






# Filter data for global trends
global_cost_data = cost_data[cost_data['Area'] == 'World']

# Prepare data for linear regression
X = global_cost_data['Year'].values.reshape(-1, 1)  # Independent variable (Year)
y = global_cost_data['Value'].values  # Dependent variable (Cost)

# Perform linear regression
model = LinearRegression()
model.fit(X, y)

# Predict values using the model
predicted_values = model.predict(X)

# Plot the cost of a healthy diet over the years
plt.figure(figsize=(10, 6))
plt.plot(global_cost_data['Year'], global_cost_data['Value'], marker='o', linestyle='-', linewidth=2, label="Actual Values")
plt.plot(global_cost_data['Year'], predicted_values, color='red', linestyle='--', label="Linear Regression")
plt.title("Global Cost of a Healthy Diet Over 5 Years", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Cost (PPP dollars per person per day)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print regression details
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```
