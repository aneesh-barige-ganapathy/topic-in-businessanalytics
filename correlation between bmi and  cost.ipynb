```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
file1 = pd.read_csv(r"C:\Users\bgane\Desktop\TOPICS DATA\data\cleaned_merged_data.csv")  # File with BMI and production data
file2 = pd.read_csv(r"C:\Users\bgane\Downloads\cost over 5 years.csv")  # File with cost data

# Merge the datasets on common columns
merged_data = pd.merge(file1, file2, on=['Year'])

# Select relevant columns for correlation
columns_of_interest = ['BMI Value', 'Value']  # Replace 'Value' with the specific cost column name
correlation_data = merged_data[columns_of_interest]

# Drop rows with missing values
correlation_data = correlation_data.dropna()

# Calculate the correlation coefficient
spearman_corr = correlation_data.corr(method='spearman').iloc[0, 1]
print(f"Spearman Correlation between BMI and Cost of Healthy Diet: {spearman_corr:.2f}")


# Visualize the correlation
sns.scatterplot(x='Value', y='BMI Value', data=correlation_data)
plt.title('Correlation Between BMI Value and Cost of Healthy Diet')
plt.xlabel('Cost of Healthy Diet (PPP dollars per person per day)')
plt.ylabel('BMI Value')
plt.show()
```
